```python
# California House Prices: Comprehensive Descriptive Analysis
# A Didactic Tutorial for Data Science Students

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing
from scipy import stats
import matplotlib.ticker as mtick

# Set visualization styles for better readability
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette('viridis')
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12

print("Descriptive Data Analysis of the California Housing Dataset")
print("==========================================================\n")

# 1. Data Acquisition
print("1. DATA ACQUISITION")
print("-----------------\n")

# Load the California Housing dataset from scikit-learn
california = fetch_california_housing()
df = pd.DataFrame(california.data, columns=california.feature_names)
df['MedHouseValue'] = california.target

print(f"Dataset loaded successfully with {df.shape[0]} rows and {df.shape[1]} columns.\n")

# Print feature descriptions
print("Feature Descriptions:")
feature_descriptions = {
    'MedInc': 'Median income in block group (in tens of thousands of US Dollars)',
    'HouseAge': 'Median house age in block group (in years)',
    'AveRooms': 'Average number of rooms per household',
    'AveBedrms': 'Average number of bedrooms per household',
    'Population': 'Block group population',
    'AveOccup': 'Average number of household members',
    'Latitude': 'Block group latitude',
    'Longitude': 'Block group longitude',
    'MedHouseValue': 'Median house value (in hundreds of thousands of US Dollars)'
}
for feature, description in feature_descriptions.items():
    print(f"- {feature}: {description}")
print("\n")

# 2. Data Exploration - First Look
print("2. DATA EXPLORATION - FIRST LOOK")
print("------------------------------\n")

print("First 5 rows of the dataset:")
print(df.head())
print("\n")

# Basic information about the dataset
print("Basic information about the dataset:")
print(f"Number of samples: {df.shape[0]}")
print(f"Number of features: {df.shape[1]}")
print(f"Features: {', '.join(df.columns)}")
print("\n")

# Check data types
print("Data types of each column:")
print(df.dtypes)
print("\n")

# Check for missing values
print("Missing values in each column:")
print(df.isnull().sum())
print("\n")

# 3. Descriptive Statistics
print("3. DESCRIPTIVE STATISTICS")
print("-----------------------\n")

# Generate descriptive statistics
print("Summary statistics for all features:")
print(df.describe())
print("\n")

# Detailed analysis of house values
print("Detailed analysis of Median House Values (in $100,000s):")
house_vals = df['MedHouseValue']
print(f"Mean: ${house_vals.mean():.4f} (x $100,000)")
print(f"Median: ${house_vals.median():.4f} (x $100,000)")
print(f"Mode: ${house_vals.mode()[0]:.4f} (x $100,000)")
print(f"Min: ${house_vals.min():.4f} (x $100,000)")
print(f"Max: ${house_vals.max():.4f} (x $100,000)")
print(f"Range: ${house_vals.max() - house_vals.min():.4f} (x $100,000)")
print(f"Variance: {house_vals.var():.4f}")
print(f"Standard deviation: {house_vals.std():.4f}")
print(f"Interquartile range (IQR): {house_vals.quantile(0.75) - house_vals.quantile(0.25):.4f}")
print(f"Skewness: {house_vals.skew():.4f}")
print(f"Kurtosis: {house_vals.kurtosis():.4f}")
print("\n")

# 4. Distribution Analysis
print("4. DISTRIBUTION ANALYSIS")
print("-----------------------\n")

# Create histograms for each numerical feature
plt.figure(figsize=(15, 12))
plt.suptitle('Distribution of Features in California Housing Dataset', fontsize=16)

for i, feature in enumerate(df.columns):
    plt.subplot(3, 3, i+1)
    sns.histplot(df[feature], kde=True)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    
    # Add mean and median lines
    plt.axvline(df[feature].mean(), color='red', linestyle='--', label=f'Mean: {df[feature].mean():.2f}')
    plt.axvline(df[feature].median(), color='green', linestyle='-.', label=f'Median: {df[feature].median():.2f}')
    plt.legend(fontsize=8)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()

# 5. Geographic Distribution
print("5. GEOGRAPHIC DISTRIBUTION")
print("------------------------\n")

# Create a scatter plot of houses by longitude and latitude, colored by house value
plt.figure(figsize=(12, 10))
scatter = plt.scatter(df['Longitude'], df['Latitude'], 
                     c=df['MedHouseValue'], cmap='viridis', 
                     s=10, alpha=0.7, edgecolors='none')
plt.colorbar(scatter, label='Median House Value ($100,000s)')
plt.title('Geographic Distribution of House Prices in California')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()

# 6. Income vs. House Value Analysis
print("6. INCOME VS. HOUSE VALUE ANALYSIS")
print("-------------------------------\n")

plt.figure(figsize=(10, 8))
plt.scatter(df['MedInc'], df['MedHouseValue'], alpha=0.5)
plt.title('Median Income vs. Median House Value')
plt.xlabel('Median Income (tens of thousands of US Dollars)')
plt.ylabel('Median House Value ($100,000s)')
plt.grid(True)

# Add regression line
z = np.polyfit(df['MedInc'], df['MedHouseValue'], 1)
p = np.poly1d(z)
plt.plot(df['MedInc'], p(df['MedInc']), "r--", linewidth=2)

# Add correlation coefficient to the plot
correlation = df['MedInc'].corr(df['MedHouseValue'])
plt.annotate(f'Correlation: {correlation:.4f}', 
             xy=(0.05, 0.95), xycoords='axes fraction',
             fontsize=12, ha='left', va='top',
             bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.3))

plt.show()

# 7. Correlation Analysis
print("7. CORRELATION ANALYSIS")
print("---------------------\n")

# Calculate correlation matrix
correlation_matrix = df.corr()
print("Correlation matrix between features:")
print(correlation_matrix)
print("\n")

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, fmt='.2f')
plt.title('Correlation Matrix of California Housing Features')
plt.tight_layout()
plt.show()

# 8. Identifying Significant Relationships
print("8. IDENTIFYING SIGNIFICANT RELATIONSHIPS")
print("------------------------------------\n")

# Identify the top 3 most correlated features with house value
correlations_with_target = correlation_matrix['MedHouseValue'].sort_values(ascending=False)
print("Top correlated features with Median House Value:")
print(correlations_with_target[1:4])  # Skip the first one (correlation with itself)
print("\n")

# 9. Location Impact Analysis
print("9. LOCATION IMPACT ANALYSIS")
print("-------------------------\n")

# Bin the data by location (coastal vs. inland)
df['IsCoastal'] = (df['Longitude'] <= -122) | ((df['Longitude'] <= -118) & (df['Latitude'] >= 33.5))

print("Comparison of house values between coastal and inland locations:")
coastal_stats = df.groupby('IsCoastal')['MedHouseValue'].describe()
print(coastal_stats)
print("\n")

# Visualize the difference
plt.figure(figsize=(10, 6))
sns.boxplot(x='IsCoastal', y='MedHouseValue', data=df)
plt.title('House Values: Coastal vs Inland Locations')
plt.xlabel('Is Coastal Location (True/False)')
plt.ylabel('Median House Value ($100,000s)')
plt.grid(True, axis='y')
plt.show()

# 10. Population Density Analysis
print("10. POPULATION DENSITY ANALYSIS")
print("----------------------------\n")

# Create a new feature for population density
df['PopulationDensity'] = df['Population'] / df['AveRooms']

# Creating bins for population density
density_bins = pd.qcut(df['PopulationDensity'], 4, labels=['Low', 'Medium-Low', 'Medium-High', 'High'])
df['DensityCategory'] = density_bins

# Group by density category and calculate statistics
density_stats = df.groupby('DensityCategory')['MedHouseValue'].describe()
print("House values by population density category:")
print(density_stats)
print("\n")

# Visualization
plt.figure(figsize=(10, 6))
sns.boxplot(x='DensityCategory', y='MedHouseValue', data=df)
plt.title('House Values by Population Density')
plt.xlabel('Population Density Category')
plt.ylabel('Median House Value ($100,000s)')
plt.grid(True, axis='y')
plt.show()

# 11. Ocean Proximity Effect
print("11. HOUSING AGE ANALYSIS")
print("--------------------\n")

# Creating bins for housing age
age_bins = pd.cut(df['HouseAge'], bins=[0, 10, 20, 30, 40, 50], labels=['<10 years', '10-20 years', '20-30 years', '30-40 years', '40+ years'])
df['AgeCategory'] = age_bins

# Group by age category and calculate statistics
age_stats = df.groupby('AgeCategory')['MedHouseValue'].describe()
print("House values by age category:")
print(age_stats)
print("\n")

# Visualization
plt.figure(figsize=(10, 6))
sns.boxplot(x='AgeCategory', y='MedHouseValue', data=df)
plt.title('House Values by Age Category')
plt.xlabel('House Age')
plt.ylabel('Median House Value ($100,000s)')
plt.grid(True, axis='y')
plt.show()

# 12. Income to House Value Ratio Analysis
print("12. INCOME TO HOUSE VALUE RATIO ANALYSIS")
print("-------------------------------------\n")

# Create income to house value ratio
df['IncomeToValueRatio'] = df['MedInc'] / df['MedHouseValue']
print("Income to House Value Ratio Statistics:")
print(df['IncomeToValueRatio'].describe())
print("\n")

# Create affordability index categories
affordability_bins = pd.qcut(df['IncomeToValueRatio'], 4, labels=['Least Affordable', 'Less Affordable', 'More Affordable', 'Most Affordable'])
df['AffordabilityCategory'] = affordability_bins

# Group by affordability category and calculate statistics
affordability_stats = df.groupby('AffordabilityCategory')[['MedInc', 'MedHouseValue', 'IncomeToValueRatio']].mean()
print("Average income and house value by affordability category:")
print(affordability_stats)
print("\n")

# Visualization
plt.figure(figsize=(10, 6))
sns.barplot(x=affordability_stats.index, y=affordability_stats['IncomeToValueRatio'])
plt.title('Income to House Value Ratio by Affordability Category')
plt.xlabel('Affordability Category')
plt.ylabel('Average Income to House Value Ratio')
plt.grid(True, axis='y')
plt.show()

# 13. Statistical Tests
print("13. STATISTICAL TESTS")
print("------------------\n")

# T-test for house value between coastal and inland areas
coastal_values = df[df['IsCoastal']]['MedHouseValue']
inland_values = df[~df['IsCoastal']]['MedHouseValue']
t_stat, p_val = stats.ttest_ind(coastal_values, inland_values, equal_var=False)
print(f"T-test for house values (coastal vs. inland):")
print(f"t-statistic: {t_stat:.4f}")
print(f"p-value: {p_val:.8f}")
print(f"Conclusion: {'Significant difference' if p_val < 0.05 else 'No significant difference'} at 5% level\n")

# ANOVA for house values across different age categories
age_groups = [df[df['AgeCategory'] == category]['MedHouseValue'] for category in df['AgeCategory'].unique() if not pd.isna(category)]
age_categories = [category for category in df['AgeCategory'].unique() if not pd.isna(category)]
f_stat, p_val = stats.f_oneway(*age_groups)
print(f"ANOVA for house values across age categories:")
print(f"f-statistic: {f_stat:.4f}")
print(f"p-value: {p_val:.8f}")
print(f"Conclusion: {'Significant difference' if p_val < 0.05 else 'No significant difference'} at 5% level\n")

# 14. Summary of Findings
print("14. SUMMARY OF FINDINGS")
print("--------------------\n")

print("Key observations from the descriptive analysis:")
print("1. The dataset contains information on housing districts in California.")
print("2. The median house value ranges from $0.15 million to $5.00 million.")
print("3. Median income is the strongest predictor of house value (correlation: 0.68).")
print("4. Coastal areas tend to have higher house values than inland areas.")
print("5. Houses in areas with high population density generally have higher values.")
print("6. Newer houses tend to have higher values than older houses.")
print("7. There is a statistically significant difference in house values between coastal and inland areas.")
print("8. The affordability (income to house value ratio) varies significantly across California.")
print("\n")

print("CONCLUSION:")
print("This descriptive analysis provides insights into the factors affecting house prices in California.")
print("Income, location, population density, and house age all play important roles in determining house values.")
print("This dataset is ideal for regression tasks to predict house prices based on these features.")
```
