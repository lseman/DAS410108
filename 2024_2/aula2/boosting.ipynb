```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor

# Generate synthetic data
np.random.seed(42)
X = np.linspace(0, 10, 100).reshape(-1, 1)
y = np.sin(X).flatten() + np.random.normal(scale=0.1, size=X.flatten().shape)

# Plot the initial data
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='True Data')
plt.title("True Data")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()

# Initial prediction (mean)
y_pred = np.full_like(y, y.mean())

# Plot the initial prediction
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='True Data')
plt.plot(X, y_pred, color='red', label='Initial Prediction (Mean)')
plt.title("Initial Prediction")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()

# Calculate residuals
residuals = y - y_pred

# Train first tree on residuals
tree_1 = DecisionTreeRegressor(max_depth=2, random_state=42)
tree_1.fit(X, residuals)
y_pred_1 = tree_1.predict(X)

# Update predictions
y_pred += y_pred_1

# Plot the first tree's prediction
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='True Data')
plt.plot(X, y_pred, color='red', label='Updated Prediction (After 1st Tree)')
plt.title("After 1st Tree")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()

# Calculate new residuals
residuals = y - y_pred

# Train second tree on residuals
tree_2 = DecisionTreeRegressor(max_depth=2, random_state=42)
tree_2.fit(X, residuals)
y_pred_2 = tree_2.predict(X)

# Update predictions
y_pred += y_pred_2

# Plot the second tree's prediction
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='True Data')
plt.plot(X, y_pred, color='red', label='Updated Prediction (After 2nd Tree)')
plt.title("After 2nd Tree")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()

# Final prediction after 3 trees
residuals = y - y_pred
tree_3 = DecisionTreeRegressor(max_depth=2, random_state=42)
tree_3.fit(X, residuals)
y_pred_3 = tree_3.predict(X)
y_pred += y_pred_3

# Plot the final prediction
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='True Data')
plt.plot(X, y_pred, color='red', label='Final Prediction (After 3rd Tree)')
plt.title("Final Model After 3 Trees")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
