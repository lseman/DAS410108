```python
# Descriptive Data Analysis with Python
# A Comprehensive Tutorial for Students

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Set visualization styles
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette('viridis')
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12

print("Descriptive Data Analysis of the Iris Dataset")
print("=============================================\n")

# 1. Data Acquisition
print("1. DATA ACQUISITION")
print("-----------------\n")

# Load the Iris dataset from seaborn
iris = sns.load_dataset('iris')

print(f"Dataset loaded successfully with {iris.shape[0]} rows and {iris.shape[1]} columns.\n")

# 2. Data Exploration - First Look
print("2. DATA EXPLORATION - FIRST LOOK")
print("------------------------------\n")

print("First 5 rows of the dataset:")
print(iris.head())
print("\n")

# Basic information about the dataset
print("Basic information about the dataset:")
print(f"Number of samples: {iris.shape[0]}")
print(f"Number of features: {iris.shape[1]}")
print(f"Features: {', '.join(iris.columns)}")
print(f"Target classes: {', '.join(iris['species'].unique())}")
print("\n")

# Check data types
print("Data types of each column:")
print(iris.dtypes)
print("\n")

# Check for missing values
print("Missing values in each column:")
print(iris.isnull().sum())
print("\n")

# 3. Descriptive Statistics
print("3. DESCRIPTIVE STATISTICS")
print("-----------------------\n")

# Generate descriptive statistics
print("Summary statistics for numerical features:")
print(iris.describe())
print("\n")

# Distribution by species
print("Distribution of samples by species:")
species_counts = iris['species'].value_counts()
print(species_counts)
print(f"Percentage: {100 * species_counts / len(iris)}")
print("\n")

# 4. Distribution Analysis
print("4. DISTRIBUTION ANALYSIS")
print("-----------------------\n")

# Create histograms for each feature
plt.figure(figsize=(14, 10))
plt.suptitle('Distribution of Features', fontsize=16)

for i, feature in enumerate(iris.columns[:-1]):
    plt.subplot(2, 2, i+1)
    sns.histplot(data=iris, x=feature, kde=True)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    
    # Add mean and median lines
    plt.axvline(iris[feature].mean(), color='red', linestyle='--', label=f'Mean: {iris[feature].mean():.2f}')
    plt.axvline(iris[feature].median(), color='green', linestyle='-.', label=f'Median: {iris[feature].median():.2f}')
    plt.legend()

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()

# 5. Boxplots for Features by Species
print("5. BOXPLOT ANALYSIS BY SPECIES")
print("----------------------------\n")

plt.figure(figsize=(14, 10))
plt.suptitle('Distribution of Features by Species', fontsize=16)

for i, feature in enumerate(iris.columns[:-1]):
    plt.subplot(2, 2, i+1)
    sns.boxplot(x='species', y=feature, data=iris)
    plt.title(f'{feature} by Species')
    plt.xticks(rotation=45)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()

# 6. Calculate and Interpret Measures of Central Tendency
print("6. MEASURES OF CENTRAL TENDENCY")
print("-----------------------------\n")

print("Calculating measures of central tendency for each feature by species:")
central_tendency = iris.groupby('species').agg({
    'sepal_length': ['mean', 'median', 'min', 'max'],
    'sepal_width': ['mean', 'median', 'min', 'max'],
    'petal_length': ['mean', 'median', 'min', 'max'],
    'petal_width': ['mean', 'median', 'min', 'max']
})
print(central_tendency)
print("\n")

# 7. Calculate and Interpret Measures of Dispersion
print("7. MEASURES OF DISPERSION")
print("-----------------------\n")

print("Calculating measures of dispersion for each feature by species:")
dispersion = iris.groupby('species').agg({
    'sepal_length': ['std', 'var', 'sem', lambda x: x.quantile(0.75) - x.quantile(0.25)],
    'sepal_width': ['std', 'var', 'sem', lambda x: x.quantile(0.75) - x.quantile(0.25)],
    'petal_length': ['std', 'var', 'sem', lambda x: x.quantile(0.75) - x.quantile(0.25)],
    'petal_width': ['std', 'var', 'sem', lambda x: x.quantile(0.75) - x.quantile(0.25)]
})
dispersion.columns = dispersion.columns.map(lambda x: f"{x[0]}_{x[1]}" if x[1] != '<lambda_0>' else f"{x[0]}_IQR")
print(dispersion)
print("\n")

# 8. Skewness and Kurtosis
print("8. SKEWNESS AND KURTOSIS")
print("-----------------------\n")

print("Skewness for each feature (by species):")
skewness = iris.groupby('species').agg({
    'sepal_length': lambda x: stats.skew(x),
    'sepal_width': lambda x: stats.skew(x),
    'petal_length': lambda x: stats.skew(x),
    'petal_width': lambda x: stats.skew(x)
})
skewness.columns = ['sepal_length_skewness', 'sepal_width_skewness', 'petal_length_skewness', 'petal_width_skewness']
print(skewness)
print("\n")

print("Kurtosis for each feature (by species):")
kurtosis = iris.groupby('species').agg({
    'sepal_length': lambda x: stats.kurtosis(x),
    'sepal_width': lambda x: stats.kurtosis(x),
    'petal_length': lambda x: stats.kurtosis(x),
    'petal_width': lambda x: stats.kurtosis(x)
})
kurtosis.columns = ['sepal_length_kurtosis', 'sepal_width_kurtosis', 'petal_length_kurtosis', 'petal_width_kurtosis']
print(kurtosis)
print("\n")

# 9. Correlation Analysis
print("9. CORRELATION ANALYSIS")
print("---------------------\n")

# Calculate correlation matrix
correlation_matrix = iris.drop('species', axis=1).corr()
print("Correlation matrix between features:")
print(correlation_matrix)
print("\n")

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Iris Features')
plt.show()

# 10. Relationship Visualization
print("10. RELATIONSHIP VISUALIZATION")
print("---------------------------\n")

# Pairplot for visual correlation analysis
plt.figure(figsize=(12, 10))
sns.pairplot(iris, hue='species', markers=['o', 's', 'D'])
plt.suptitle('Pairwise Relationships Between Features by Species', y=1.02, fontsize=16)
plt.show()

# 11. Feature Comparisons
print("11. FEATURE COMPARISONS")
print("---------------------\n")

# Comparing petal length to petal width
plt.figure(figsize=(10, 8))
sns.scatterplot(x='petal_length', y='petal_width', hue='species', style='species', data=iris)
plt.title('Petal Width vs. Petal Length by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')
plt.show()

# 12. Statistical Tests
print("12. STATISTICAL TESTS")
print("------------------\n")

print("T-tests for each feature comparing setosa vs. versicolor:")
for feature in iris.columns[:-1]:
    setosa = iris[iris['species'] == 'setosa'][feature]
    versicolor = iris[iris['species'] == 'versicolor'][feature]
    t_stat, p_val = stats.ttest_ind(setosa, versicolor, equal_var=False)
    print(f"{feature}: t-statistic = {t_stat:.4f}, p-value = {p_val:.8f}")
print("\n")

# 13. Summary of Findings
print("13. SUMMARY OF FINDINGS")
print("--------------------\n")

print("Key observations from the descriptive analysis:")
print("1. The dataset contains 150 samples, with 50 samples from each of the three species.")
print("2. There are no missing values in the dataset.")
print("3. Setosa species is clearly distinguishable from the other two species.")
print("4. Petal length and petal width show the strongest correlation.")
print("5. Species show significant differences in their measurements (confirmed by t-tests).")
print("6. Iris setosa has the smallest petals, while Iris virginica has the largest.")
print("7. The distributions of petal measurements show clear separation between species.")
print("\n")

print("CONCLUSION:")
print("This descriptive analysis provides insights into the characteristics of three Iris species.")
print("The measurements, particularly of the petals, can be used to differentiate between the species.")
print("This makes the Iris dataset an excellent example for classification tasks in machine learning.")
```
