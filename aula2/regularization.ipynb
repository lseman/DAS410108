```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge, Lasso
from sklearn.datasets import make_regression

# Generate synthetic data
X, y = make_regression(n_samples=100, n_features=20, noise=0.1, random_state=42)

# Introduce some sparsity in the coefficients for clarity
true_coefs = np.random.randn(20)
true_coefs[true_coefs < 0.5] = 0
y = np.dot(X, true_coefs) + np.random.normal(scale=0.1, size=y.shape)

# Perform Ridge regression
ridge = Ridge(alpha=1.0)
ridge.fit(X, y)
ridge_coefs = ridge.coef_

# Perform Lasso regression
lasso = Lasso(alpha=0.1)
lasso.fit(X, y)
lasso_coefs = lasso.coef_

# Plot the coefficients for comparison
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.bar(range(len(ridge_coefs)), ridge_coefs, color='blue')
plt.title("Ridge Regression Coefficients")
plt.xlabel("Feature Index")
plt.ylabel("Coefficient Value")

plt.subplot(1, 2, 2)
plt.bar(range(len(lasso_coefs)), lasso_coefs, color='red')
plt.title("Lasso Regression Coefficients")
plt.xlabel("Feature Index")
plt.ylabel("Coefficient Value")

plt.tight_layout()
plt.show()
```
