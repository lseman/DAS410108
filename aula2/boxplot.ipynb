```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(10)
data = np.random.normal(100, 20, 200)

# Create a box plot
fig, ax = plt.subplots(figsize=(10, 6))
box = ax.boxplot(data, vert=False, patch_artist=True, showmeans=True, widths=0.7)

# Customizing the plot
colors = ['lightblue']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

# Adding text annotations
median = np.median(data)
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1
lower_whisker = q1 - 1.5 * iqr
upper_whisker = q3 + 1.5 * iqr

# Annotating the median
ax.annotate(f'Median (Q2): {median:.2f}', xy=(median, 1), xytext=(median, 1.2),
            arrowprops=dict(facecolor='black', arrowstyle="->"),
            ha='center')

# Annotating Q1 and Q3
ax.annotate(f'Q1 (25th percentile): {q1:.2f}', xy=(q1, 1), xytext=(q1-20, 1.3),
            arrowprops=dict(facecolor='black', arrowstyle="->"),
            ha='center')

ax.annotate(f'Q3 (75th percentile): {q3:.2f}', xy=(q3, 1), xytext=(q3+20, 1.3),
            arrowprops=dict(facecolor='black', arrowstyle="->"),
            ha='center')

# Annotating the IQR
ax.annotate(f'IQR: {iqr:.2f}', xy=((q1 + q3) / 2, 1), xytext=((q1 + q3) / 2, 1.5),
            arrowprops=dict(facecolor='black', arrowstyle="->"),
            ha='center')

# Annotating the whiskers
ax.annotate(f'Lower whisker (1.5 * IQR): {lower_whisker:.2f}', xy=(lower_whisker, 1), xytext=(lower_whisker-30, 0.8),
            arrowprops=dict(facecolor='black', arrowstyle="->"),
            ha='center')

ax.annotate(f'Upper whisker (1.5 * IQR): {upper_whisker:.2f}', xy=(upper_whisker, 1), xytext=(upper_whisker+30, 0.8),
            arrowprops=dict(facecolor='black', arrowstyle="->"),
            ha='center')

# Annotating outliers
for outlier in box['fliers'][0].get_xdata():
    ax.annotate(f'Outlier: {outlier:.2f}', xy=(outlier, 1), xytext=(outlier, 1.5),
                ha='center')

# Title and labels
ax.set_title('Box Plot with Annotations')
ax.set_xlabel('Value')
ax.set_yticks([])

# Show the plot
plt.show()
```
